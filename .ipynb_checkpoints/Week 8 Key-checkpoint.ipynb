## Swiss Rolls and SVCs

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import manifold, datasets, svm
```

```python
Roll, color = datasets.samples_generator.make_swiss_roll(n_samples=1000, noise=0.05)
Roll.shape
```

```python
color.shape
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.neighbors import kneighbors_graph
connectivity = kneighbors_graph(Roll, n_neighbors=5, include_self=False)

ward = AgglomerativeClustering(n_clusters=4, connectivity=connectivity,
                               linkage='ward').fit(Roll)

point_ID = ward.labels_
#point_ID = numpy.array(point_ID)
point_ID = point_ID + 1
np.unique(point_ID)
print (point_ID)
```

```python
#Roll[:, 1] *= .5
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
#play around with viewing in 3D
ax.view_init(15, -85)
color = ['r','b','g','c','m','k']
for i in np.arange(1,5):
    ax.scatter(Roll[point_ID==i,0],Roll[point_ID==i,1],Roll[point_ID==i,2], c=color[i-1], marker='o')

plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.neighbors import kneighbors_graph
connectivity = kneighbors_graph(Roll, n_neighbors=3, include_self=False)

ward = AgglomerativeClustering(n_clusters=6, connectivity=connectivity,
                               linkage='ward').fit(Roll)

point_ID = ward.labels_
#point_ID = numpy.array(point_ID)
point_ID = point_ID + 1
np.unique(point_ID)
print (point_ID)
```

```python
train_ind = np.random.choice(np.arange(0,len(point_ID)), size=(len(point_ID)*0.9), replace=False, p=None)


len(train_ind) == len(set(train_ind))
```

```python
test_ind = np.delete(np.arange(0,len(point_ID)), train_ind)
len(test_ind)
```

```python


from sklearn.svm import SVC
identify = SVC()
identify.fit(Roll[train_ind,:], point_ID[train_ind]) 

predicted = identify.predict(Roll[test_ind,:])
actual = point_ID[test_ind]
predicted == actual
```

```python
# Author: Fabian Pedregosa -- <fabian.pedregosa@inria.fr>
# License: BSD 3 clause (C) INRIA 2011

print(__doc__)

import matplotlib.pyplot as plt

# This import is needed to modify the way figure behaves
from mpl_toolkits.mplot3d import Axes3D
Axes3D

#----------------------------------------------------------------------
# Locally linear embedding of the swiss roll

from sklearn import manifold, datasets
X, color = datasets.samples_generator.make_swiss_roll(n_samples=1500)

print("Computing LLE embedding")
X_r, err = manifold.locally_linear_embedding(X, n_neighbors=12,
                                             n_components=2)
print("Done. Reconstruction error: %g" % err)

#----------------------------------------------------------------------
# Plot result

fig = plt.figure()
try:
    # compatibility matplotlib < 1.0
    ax = fig.add_subplot(211, projection='3d')
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.Spectral)
except:
    ax = fig.add_subplot(211)
    ax.scatter(X[:, 0], X[:, 2], c=color, cmap=plt.cm.Spectral)

ax.set_title("Original data")
ax = fig.add_subplot(212)
ax.scatter(X_r[:, 0], X_r[:, 1], c=color, cmap=plt.cm.Spectral)
plt.axis('tight')
plt.xticks([]), plt.yticks([])
plt.title('Projected data')
plt.show()
```
